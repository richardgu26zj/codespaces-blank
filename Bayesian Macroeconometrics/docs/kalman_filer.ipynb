**Kaman filter Note**

The state space model is defined as follows:

$y_t = H s_t + v_t\,,\quad v_t\sim\mathcal{N}(0, R)$

$s_t = A s_{t-1} + w_t\,,\quad w_t \sim\mathcal{N}(0,Q)$

The first step is prediction. Define $s_{t|t-1} = \mathbb{E}[s_t|y_{1:t-1}]$ and $P_{t|t-1} = \mathbb{E}[(s_t-s_{t|t-1})(s_t-s{t|t-1})']$, and we have 

$$
s_{t|t-1} = A s_{t-1|t-1}\,,
$$

$$
\begin{align*}
P_{t|t-1}  & = \mathbb{E}[(s_t - s_{t|t-1})(s_t - s_{t|t-1})']\,,\\
               &  = \mathbb{E}[(As_{t-1} - As_{t-1|t-1} + w_t)(As_{t-1} - As_{t-1|t-1} + w_t)']\,,\\
               & = \mathbb{E}[A(s_{t-1}-s_{t-1|t-1})(s_{t-1}-s_{t-1|t-1})'A' + w_tw_t']\,,\\
               & = AP_{t-1|t-1}A' + Q\,,
\end{align*}
$$
